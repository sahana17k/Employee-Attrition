```python
#Finding outliers

f,a=plt.subplots(3,3,figsize=(20,10),dpi=100)
num=0
for i in range(3):
    for j in range(3):
        sns.boxplot(y=Data_copy_x[true_numeric[num]],ax=a[i,j])
        num+=1
        
plt.show()        
```

```python
#This shows us that we need to do outlier treatment for our true_numeric category
```

```python
Data_copy_x.describe()
```

```python
#Treating outliers

def outlier_treatment(col):
    
    Q3 = Data_copy_x[true_numeric].quantile(0.75)
    Q1 = Data_copy_x[true_numeric].quantile(0.25)
    IQR = Q3-Q1

    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR

    Data_copy_x[true_numeric] = np.where(Data_copy_x[true_numeric]>upper_limit,upper_limit,
                             np.where(Data_copy_x[true_numeric]<lower_limit,lower_limit,
                                     Data_copy_x[true_numeric]))

```

```python
for i in true_numeric:
    outlier_treatment(i)
```

```python
f,a=plt.subplots(3,3,figsize=(20,10),dpi=100)
num=0
for i in range(3):
    for j in range(3):
        sns.boxplot(y=Data_copy_x[true_numeric[num]],ax=a[i,j])
        num+=1
        
plt.show()        
```
